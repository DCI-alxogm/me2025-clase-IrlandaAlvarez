```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
from scipy.optimize import minimize
```

```python
x,y,yerr=np.loadtxt("/content/sample_data/data_MN.dat",skiprows=1).T
```

```python
y=y+np.random.randint(-2,1,len(y))
```

```python
yerr=yerr*10
```

```python
def R(params):
  m=params[0]
  b=params[1]
  residuo=np.sum((y-m*x-b)**2)
  return residuo

def Chi2(params):
  m=params[0]
  b=params[1]
  residuo=np.sum((y-m*x-b)**2//yerr**2)
  return residuo
```

```python
params_0=[0,0]
bounds=[(0,10),(0,10)]
res=minimize(R,params_0,bounds=bounds)
m_r,b_r=res.x
m_r,b_r
```

```python
params_0=[0,0]
bounds=[(0,10),(0,10)]
res=minimize(Chi2,params_0,bounds=bounds)
m_chi,b_chi=res.x
m_chi,b_chi
```

```python
plt.errorbar(x,y,yerr=yerr,fmt='.')
plt.errorbar(x,m_r*x+b_r)
```
